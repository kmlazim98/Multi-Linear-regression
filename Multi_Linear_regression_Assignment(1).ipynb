```python
import pandas as pd
start = pd.read_csv('50_Startups.csv')
start = start.rename({'R&D Spend':'RnDSpend', 'Marketing Spend':'MarketingSpend'},axis = 1)
start['TotalSpend'] = start['RnDSpend']+start['Administration']+start['MarketingSpend']
start
```

```python
start.corr()
```

Marketing Spend has minimal ifluence on Profit so we can neglet it

```python
start = start.drop('MarketingSpend',axis = 1)
start
```

```python
statewise = start.groupby('State').mean()
statewise
```

```python
import matplotlib.pyplot as plt
plt.bar(statewise.index,statewise['Profit'])
plt.title('Average Profit in deferent cities')
plt.show
```

```python

plt.subplots(1,1,figsize = (10,5))
plt.title('Relationship between deferent expenses and profit')
plt.plot(start.index,start['Profit'],color = 'blue')
plt.plot(start.index,start['RnDSpend'],color = 'red')
plt.plot(start.index,start['Administration'],color = 'green')

plt.subplots(1,1,figsize = (10,5))
plt.title('Relationship between Administration Spend and Profit')
plt.plot(start.index,start['Profit'],color = 'blue')
plt.plot(start.index,start['Administration'],color = 'red')

plt.subplots(1,1,figsize = (10,5))
plt.title('Relationship between Total Spend and Profit')
plt.plot(start.index,start['TotalSpend'],color = 'red')
plt.plot(start.index,start['Profit'],color = 'blue')

plt.subplots(1,1,figsize = (10,5))
plt.title('Relationship between RnD Spend and Profit')
plt.plot(start.index,start['RnDSpend'],color = 'red')
plt.plot(start.index,start['Profit'],color = 'blue')


plt.show()
```

```python
import statsmodels.formula.api as smf
model = smf.ols("Profit ~ RnDSpend + Administration", data = start).fit()
model.params
```

```python
start['Prediction'] = model.predict(start)
start
```

```python
model.summary()
```

# prepare a prediction model for predicting Price.

```python
car = pd.read_csv('ToyotaCorolla.csv', encoding = 'latin1')
car
```

```python
car = car.drop('Id', axis = 1)
car
```

```python
corr = car.corr()
abs(corr['Price']).sort_values(ascending = False)
```

We can neglet columns with low correlation to profit

```python
car = car[['Price','Mfg_Year','Boardcomputer','Automatic_airco','Weight','CD_Player','Airco','Age_08_04','KM']]
car
```

```python
corr = car.corr()
corr
```

We can neglet features with high correleation with other features 

```python
car = car[['Price','Boardcomputer','Automatic_airco','Weight','CD_Player','Airco','Age_08_04','KM']]
```

```python
import statsmodels.formula.api as smf
model = smf.ols("Price ~ Boardcomputer + Automatic_airco + Weight + CD_Player + Airco + Age_08_04 + KM ", data = car).fit()
model.params
```

```python
car['Prediction'] = model.predict(car)
car
```

```python
model.summary()
```
